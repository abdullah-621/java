```python
import matplotlib.pylab as plt

height = [150, 160, 170, 180, 190]
weight = [50, 60, 65, 75, 85]
plt.scatter(height, weight)
plt.title("Height vs Weight")
plt.xlabel("Height")
plt.ylabel("Weight")
plt.show()

```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
data = [12, 15, 14, 10, 20, 21, 22, 25, 30, 18, 13, 17, 50]

print(sorted(data))

# plt.boxplot(data)
plt.boxplot(data, patch_artist=True,
            boxprops=dict(facecolor="lightblue"))
plt.grid()
plt.title("Simple Boxplot")
plt.ylabel("Values")

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

data1 = np.random.normal(50, 10, 200)
data2 = np.random.normal(60, 15, 200)
data3 = np.random.normal(55, 8, 200)

plt.boxplot([data1, data2, data3])
plt.title("Multiple Boxplots")
plt.xticks([1, 2, 3], ["Group 1", "Group 2", "Group 3"])
plt.ylabel("Values")

plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.Series([10, 12, 13, 14, 500, 15, 16, 18])
sns.histplot(data, kde=True)   # Plot histogram
plt.show()

print(data.skew())  # skew > 0  right skew, skew < 0  left skew
```

```python
data_log = np.log1p(data)  # log(1 + x) to avoid log(0)
sns.histplot(data_log, kde=True)   # Plot histogram
plt.show()
```
